## BASIC KAGGLE COMMANDS
- esc+a:add new cell above
- esc+b: add new cell below
- dd: delete cell
- ctrl+enter: run current cell
- shift+enter : run current cell and move onto next cell
- esc+m: markdown
- esc+c: code
